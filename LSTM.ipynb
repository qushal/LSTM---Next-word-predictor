```python
with open("chat_data.txt", "r", encoding = "utf-8") as f:
  text_data = f.read()

print(text_data[:500])
```

```python
faqs = text_data[:5000]
faqs
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
top5 = list(tokenizer.word_index.items())[:5]
print(top5)
```

```python
total_words = len(tokenizer.word_index)
```

```python
import re

def split_on_emojis(text):
    # regex to catch emojis
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags
        u"\U00002700-\U000027BF"  # dingbats
        u"\U0001F900-\U0001F9FF"  # supplemental symbols
        u"\U00002600-\U000026FF"  # misc symbols
        "]+", flags=re.UNICODE
    )
    # split on emoji, keep only non-empty parts
    return [s.strip() for s in emoji_pattern.split(text) if s.strip()]
```

```python
input_sequences = []
for sentence in split_on_emojis(faqs):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  # we will create our training dataset from the above tokenized_sentence

  for i in range(1, len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding = 'pre')
padded_input_sequences
```

```python
X = padded_input_sequences[:, :-1]
y = padded_input_sequences[:, -1]
```

```python
X.shape
```

```python
y.shape
```

```python
#oneHotEncoding
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes = total_words+1)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import numpy as np
```

```python
model = Sequential()
model.add(Embedding(input_dim = total_words+1, output_dim = 100))
model.add(LSTM(150))
model.add(Dense(total_words+1, activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer ='adam', metrics = ['accuracy'])
model.build(input_shape = (None, 40))
```

```python
model.summary()
```

```python
model.fit(X, y, epochs = 100)
```

```python
text = 'long time'

# like if you want to predict next five words so run a loop for 5 times
for i in range(10):
  #tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  #padding
  padded_token_text = pad_sequences([token_text], maxlen = 40, padding = 'pre')
  #predict
  pos = np.argmax(model.predict(padded_token_text))

  for word, index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
```
